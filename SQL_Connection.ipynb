```python
import pandas as pd
from sqlalchemy import create_engine
import pymysql
pymysql.install_as_MySQLdb()
from mysql_conn import password

import os
```

```python
top_200_weekly = pd.read_csv('spotify-chart/top_200_weekly.csv')

top_200_weekly.drop(columns = 'Unnamed: 0', inplace = True)
# top_200_weekly.rename(columns = {'Unnamed: 0':'index'}, inplace = True)
top_200_weekly.rename(columns = {'Track Name':'Track'}, inplace = True)

top_200_weekly['ID'] = top_200_weekly['URL'].str.split('/', expand = True)[4]

top_200_weekly.head()
```

```python
connection_string = (f"root:{password}@localhost")

engine = create_engine(f"mysql://{connection_string}")
engine.execute("DROP DATABASE IF EXISTS spot_db")
engine.execute("CREATE DATABASE spot_db")
engine = create_engine(f"mysql://{connection_string}/spot_db?charset=utf8mb4")
```

```python
engine.execute("USE spot_db")
top_200_weekly.to_sql(
    name = 'top_200_weekly', con = engine,
    if_exists = 'replace', chunksize = 1)
with engine.connect() as con:
    con.execute('ALTER TABLE `top_200_weekly` ADD PRIMARY KEY (`index`);')
```
